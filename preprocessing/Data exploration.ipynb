```python
import pandas as pd
import numpy as np
```

```python
full_dataset = pd.read_excel(
    io="./data/veteran_suicides_dataset/veteran_suicides_full_dataset.xlsx",
    sheetname=None)
```

```python
full_dataset.keys()
```

```python
years = ['2005', '2006', '2007', '2008', '2009', '2010', '2011']
yearly_data = []
for year in years:
    full_dataset[year]['year'] = year
    yearly_data.append(full_dataset[year])
data = pd.concat(yearly_data)
pd.options.display.max_columns = 1000
data.head()
```

```python
data.tail()
```

```python
data = data[data['state'].notnull()]
data = data[data['state'] != 'Checked against numbers from collected datasets']
columns_to_drop = ['vet_15_24', 'vet_18_29', 'vet_17_34', 'vet_25_34', 'vet_30_39', 'vet_25_44',
                  'vet_35_44', 'vet_40_49', 'vet_35-54', 'vet_45_54', 'vet_50_59', 'vet_45_64',
                  'vet_55_64', 'vet_60', 'vet_65']
data_cleaned = data.drop(columns_to_drop, axis=1)
data_cleaned.head()
```

```python
data_cleaned.reset_index(inplace=True, drop=True)
data_cleaned.to_csv("aggregate.csv", index=False)
```

```python
data_from_file = pd.read_csv("aggregate.csv")
data_from_file.head()
```

```python
vet_pop = data_from_file[data_from_file['state'] == 'Total'][['year','vet_pop']].rename(columns={'vet_pop':'population'})
civ_pop = data_from_file[data_from_file['state'] == 'Total'][['year','overall_pop_18']].rename(columns={'overall_pop_18':'population'})
civ_pop['population'] = civ_pop['population'] - vet_pop['population']
vet_pop['Population Type'] = 'Veteran Population'
civ_pop['Population Type'] = 'Adult Civilian Population'

pop = pd.concat([vet_pop, civ_pop], axis=0)
pop.to_json('pop_data.json', orient='records')

print(np.mean(vet_pop['population']) / np.mean(civ_pop['population']))

pop
```

```python
vet_suicides = data_from_file[data_from_file['state'] == 'Total'][['year','vet_suicides']].rename(columns={'vet_suicides':'suicides'})
civ_suicides = data_from_file[data_from_file['state'] == 'Total'][['year','all_suicides']].rename(columns={'all_suicides':'suicides'})
civ_suicides['suicides'] = civ_suicides['suicides'] - vet_suicides['suicides']
vet_suicides['Suicide Type'] = 'Veteran Suicides'
civ_suicides['Suicide Type'] = 'Civilian Suicides'

suicides = pd.concat([vet_suicides, civ_suicides], axis=0)
suicides.to_json('suicides_data.json', orient='records')

print(np.mean(vet_suicides['suicides']) / np.mean(all_suicides['suicides']))

suicides
```

```python
vet_suicides = data_from_file[data_from_file['state'] == 'Total'][['year','vet_suicides']].rename(columns={'vet_suicides':'suicides'})
all_suicides = data_from_file[data_from_file['state'] == 'Total'][['year','all_suicides']].rename(columns={'all_suicides':'suicides'})
vet_suicides['Suicide Type'] = 'Veteran Suicides'
all_suicides['Suicide Type'] = 'Overall Suicides'

suicides = pd.concat([vet_suicides, all_suicides], axis=0)
suicides['suicides'] = suicides['suicides'] / 1000
suicides.to_json('suicides_data.json', orient='records')

print(np.mean(vet_suicides['suicides']) / np.mean(all_suicides['suicides']))

suicides
```

```python
rates_by_state = data_from_file[['state', 'vet_rate', 'civ_rate']]
rates_by_state['state'] = rates_by_state['state'].map(lambda state: state.split('*')[0])
state_mapping = {'Alabama' : 'AL', 'Alaska':'AK', 'Arizona':'AZ', 'Arkansas':'AR', 'California':'CA',
       'Colorado':'CO', 'Connecticut':'CT', 'Delaware':'DE', 'Florida':'FL', 'Georgia':'GA',
       'Hawaii':'HI', 'Idaho':'ID', 'Illinois':'IL', 'Indiana':'IN', 'Iowa':'IA', 'Kansas':'KS',
       'Kentucky':'KY', 'Louisiana':'LA', 'Maine':'ME', 'Maryland':'MD', 'Massachusetts':'MA',
       'Michigan':'MI', 'Minnesota':'MN', 'Mississippi':'MS', 'Missouri':'MO', 'Montana':'MT',
       'Nebraska':'NE', 'Nevada':'NV', 'New Hampshire':'NH', 'New Jersey':'NJ', 'New Mexico':'NM',
       'New York':'NY', 'North Carolina':'NC', 'North Dakota':'ND', 'Ohio':'OH', 'Oklahoma':'OK',
       'Oregon':'OR', 'Pennsylvania':'PA', 'Rhode Island':'RI', 'South Carolina':'SC',
       'South Dakota':'SD', 'Tennessee':'TN', 'Texas':'TX', 'Utah':'UT', 'Vermont':'VT', 'Virginia':'VA',
       'Washington':'WA', 'West Virginia':'WV', 'Wisconsin':'WI', 'Wyoming':'WY'}
rates_by_state['state'] = rates_by_state['state'].map(state_mapping)
pd.options.display.max_rows=400
rates_by_state
```

```python
means = rates_by_state.groupby('state', as_index=False).mean()
inv_state_mapping = {v: k for k, v in state_mapping.items()}
means['state_full'] = means['state'].map(inv_state_mapping)
means
```

```python
means.merge(states_data, on='state', how='left')
```
